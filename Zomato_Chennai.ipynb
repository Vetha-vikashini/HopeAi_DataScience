# Q1) How many restaurants are in Chennai for each type of cuisine?Â¶

```python
np.unique(df['Cuisine'])
```

```python
import ast

df['Cuisine'] = df['Cuisine'].apply(safe_eval)
df = df.explode('Cuisine')
df['Cuisine'] = df['Cuisine'].str.strip()
```

```python
def safe_eval(x):
    try:
        # If it's already a list-looking string, eval it
        if isinstance(x, str) and x.strip().startswith('['):
            return ast.literal_eval(x)
        # If it's just a plain string, wrap it in a list
        elif isinstance(x, str):
            return [x.strip()]
        else:
            return []
    except:
        return []
```

# Q2) Which places have the highest rated restaurant for each Cuisine Type in Chennai?

```python
high_rate = df[(df['Dining Rating'] > 4.5) | (df['Delivery Rating'] > 4.5)]
high_rate
```

```python
fig = px.histogram(high_rate, x='Location', color='Cuisine', title= 'No. of Best Restaurant for each Cuisine Type by Places').update_xaxes(categoryorder="total descending")
 
fig.show() 
```

# Q3) What is the Avg Price Distibution of highest rated restaurant for each Cuisine Type in Mumbai?

```python
price=high_rate.groupby(['Location','Cuisine'])['Price for 2'].mean().reset_index()
```

```python
price
```

```python
fig = px.scatter(high_rate, x="Location", y="Price for 2", color="Cuisine", symbol="Cuisine",  
           title=' Avg Price Distibution of High rated restaurant for each Cuisine Type').update_traces(marker_size=10)
 
fig.show()
```

# Q6) Find the list of Affordable Restaurants?

```python
m=df['Price for 2'].max()
aff=m/4
aff
```

```python
aff_rest_df = df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location']] 
aff_rest_df = aff_rest_df[aff_rest_df['Price for 2'] <= 1250.0] 
aff_rest_df.sort_values(by='Price for 2', inplace=True) 
aff_rest_df.rename(columns={'Name of Restaurant':'Name'},inplace=True)
```

```python
 
highrate_rest_df = df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location','Dining Rating','Delivery Rating']] 
highrate_rest_df = highrate_rest_df[(highrate_rest_df['Dining Rating'] >= 4.5) | (highrate_rest_df['Delivery Rating'] >= 4.5) ] 
highrate_rest_df.sort_values(by='Price for 2', inplace=True) 
highrate_rest_df.rename(columns={'Name of Restaurant':'Name'},inplace=True)
```

```python
highrate_rest_df.columns
```

```python
highrate_aff_df = pd.merge(aff_rest_df, highrate_rest_df, how='inner', on=['Name', 'Location']) 
highrate_aff_df
```

```python
highrate_aff_df = highrate_aff_df[['Name', 'Price for 2_x', 'Cuisine_x', 'Location']] 
highrate_aff_df.rename(columns={'Price for 2_x':'Price for 2','Cuisine_x':'Cuisine'})
```
