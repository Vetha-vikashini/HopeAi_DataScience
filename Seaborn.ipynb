```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Placement.csv')
```

```python
data
```

```python
print("No of rows in dataset:",data.shape[0])
print("No of cols in the dataset:",data.shape[1])
```

```python
data.shape
```

# univariate

```python
sns.displot(data['ssc_p'],kind='hist',rug=True,kde=True,legend=False)
```

# Bivariate

```python
# sns.jointplot(x=data['ssc_p'],y=data['hsc_p'],kind='hex')
sns.jointplot(x='ssc_p',y='hsc_p',data=data,kind='hex')
plt.show()
```

```python
sns.pairplot(data,diag_kind='kde',kind='scatter',hue='gender',palette='husl')
plt.show()
```

# Categorical 

```python
sns.stripplot(x='ssc_b',y='salary',data=data,jitter=True,color='green',edgecolor='yellow',linewidth=1)

"""To prevent data points from overlapping, stripplots add a small amount of random noise (jitter) along the categorical axis,
making it easier to see the distribution within each category.""" 
```

```python
sns.stripplot(x='ssc_b',y='salary',data=data,color='green',edgecolor='yellow',linewidth=1)
```

```python
sns.swarmplot(x='ssc_b',y='salary',data=data,color='green',hue='gender',dodge=True)
```

# Distribution 

```python
sns.violinplot(y='salary',x='gender',data=data)
```

```python
sns.violinplot(y='salary',x='specialisation',data=data,hue='gender')
```

```python
sns.boxplot(y='ssc_p',x='specialisation',data=data,hue='gender')
```

```python
sns.barplot(y='salary',x='specialisation',data=data,hue='gender')
```

```python
sns.countplot(x='specialisation',data=data,hue='gender')
```

```python
sns.pointplot(y='salary',x='specialisation',data=data)
```

```python
sns.pointplot(y='salary',x='specialisation',data=data,ci=20)
```

```python
sns.regplot(x='salary',y='ssc_p',data=data)
```

```python
sns.lmplot(x='salary',y='ssc_p',data=data,hue='gender')
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('tips')
g = sb.FacetGrid(df,col='time')
g.map(plt.hist, "tip")
plt.show()

# FacetGrid, it uses different pair of variable for each subplot.
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
g = sb.PairGrid(df)
g.map(plt.scatter);
plt.show()

#PairGrid allows us to draw a grid of subplots using the same plot type to visualize data.
```
